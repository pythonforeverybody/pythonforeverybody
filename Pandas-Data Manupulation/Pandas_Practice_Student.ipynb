### Clean the data


1.   Remove the null row
2.   Remove dollar symbol
```
dataframe.dropna(how='all')
dataframe.replace(to_replace ='\$', value = '', regex = True)
```



````python
df=df.dropna(how='all')
df.head(2)
````

````python
df=df.replace(to_replace ='\$', value = '', regex = True)
df.head(2)
````

````python
df.dtypes
````

###Review the data and change the column into appropriate format
```
dataframe[cols]=dataframe[cols].apply(pd.to_numeric, errors='coerce')
dataframe['timestamp'].astype('str').astype('datetime64[ns]')
```

````python
df['Order Date']=df['Order Date'].astype('str').astype('datetime64[ns]')
````

````python
df.dtypes
````

````python
cols=['List Price Per Unit','Purchase Price Per Unit', 'Quantity','Item Subtotal','Item Subtotal Tax','Item Total']
df[cols]=df[cols].apply(pd.to_numeric, errors='coerce')
df.head()
````

````python
df.dtypes
````
